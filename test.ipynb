```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import ExtraTreesRegressor
```

```python
impute_col = 'oulad_n_assessments'

impute_features = [
    'oulad_avg_assessment_score',
    'oulad_avg_clicks',
    'oulad_active_days',
    'registration_duration_days',
    'effort_norm'
]

impute_cols = impute_features + [impute_col]
```

```python
imputer = IterativeImputer(
    estimator = ExtraTreesRegressor(n_estimators=200,random_state=42,n_jobs=1),
    max_iter=10,
    initial_strategy='constant',
    fill_value=1,
    random_state=0
)
```

```python
df[impute_cols] = imputer.fit_transform(df[impute_cols])
```

```python
df[impute_col] = (
    df[impute_col]
    .clip(lower=1)     
    .round()
    .astype(int)
)

df[impute_col].describe()
```

```python
df.isnull().sum()
```

```python
df.drop(columns='has_reg_norm', axis=1, inplace=True)
```

```python
df.columns
```

```python
target = 'cognitive_efficiency'
id = 'student_id'
```

```python
from sklearn.model_selection import train_test_split
```

```python
students = df[id].unique()

train_pool_ids, test_ids = train_test_split(students, test_size=0.20,random_state=42)

train_pool_df = df[df[id].isin(train_pool_ids)].reset_index(drop=True)
test_df = df[df[id].isin(test_ids)].reset_index(drop=True)

print("Train pool: ", train_pool_df.shape)
print("Test set: ", test_df.shape)
```

```python
train_ids, val_ids = train_test_split(train_pool_df[id].unique(), train_size=0.80, random_state=42)

train_df = train_pool_df[train_pool_df[id].isin(train_ids)].reset_index(drop=True)
val_df = train_pool_df[train_pool_df[id].isin(val_ids)].reset_index(drop=True)

print("Train set: ", train_df.shape)
print("Validation set: ", val_df.shape)
```

```python
feature_cols = [
    c for c in df.columns
    if c not in [id, target]
]

feature_cols
```

```python
#Train Set
X_train = train_df[feature_cols]
y_train = train_df[target]

#Validation Set
X_val = val_df[feature_cols]
y_val = val_df[target]

#Test Set
X_test = test_df[feature_cols]
y_test = test_df[target]
```

```python
test_student_ids = test_df[id].values
```

```python
assert set(train_df[id]).isdisjoint(val_df[id])
assert set(train_df[id]).isdisjoint(test_df[id])
assert set(val_df[id]).isdisjoint(test_df[id])

print("No student leakage ✔")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
models = {
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Extra Tree": ExtraTreesRegressor(),
    "XGBoost": XGBRegressor(),
    "Lightgbm": LGBMRegressor()
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train, y_train)

    #Predictions
    y_train_pred = model.predict(X_train)
    y_val_pred = model.predict(X_val)

    #Training Set Performance
    model_train_mae = mean_absolute_error(y_train, y_train_pred)
    model_train_mse = mean_squared_error(y_train, y_train_pred)
    model_train_r2 = r2_score(y_train, y_train_pred)

    #Validation Set Performance
    model_val_mae = mean_absolute_error(y_val, y_val_pred)
    model_val_mse = mean_squared_error(y_val, y_val_pred)
    model_val_r2 = r2_score(y_val, y_val_pred)

    print(list(models.keys())[i])

    print("Model Performance on Training Set: ")

    print('- MAE: {:.4f}'.format(model_train_mae))
    print('- MSE: {:.4f}'.format(model_train_mse))
    print('- R2 Score: {:.4f}'.format(model_train_r2))

    print('-----------------------------------')

    print("Model Performance on Validation Set: ")

    print('- MAE: {:.4f}'.format(model_val_mae))
    print('- MSE: {:.4f}'.format(model_val_mse))
    print('- R2 Score: {:.4f}'.format(model_val_r2))

    print('='*35)
```

```python
xgb = XGBRegressor()
xgb.fit(X_train,y_train)
```

```python
y_test_pred = xgb.predict(X_test)
```

```python
test_mae = mean_absolute_error(y_test, y_test_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)

print("TEST SET PERFORMANCE")
print("- MAE :", test_mae)
print("- RMSE:", test_mse)
print("- R2  :", test_r2)
```
